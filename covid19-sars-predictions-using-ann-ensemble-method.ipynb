## XGboost has performed better than Random Forest overall,well lets see the Neural Network Implementation

```python
mm = MinMaxScaler()
```

```python
X_train_mm = mm.fit_transform(X_train)
X_test_mm = mm.transform(X_test)
```

```python
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.models import  Sequential
```

```python
model = Sequential()
model.add(Dense(128,activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(256,activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
r = model.fit(X_train_mm,y_train, epochs = 100,validation_data=(X_test_mm,y_test))
```

```python
plt.plot(r.history['loss'],label = 'loss',color = 'red')
plt.plot(r.history['val_loss'],label = 'validation_loss',color = 'blue')
plt.legend()
plt.show()
```

```python
plt.plot(r.history['accuracy'],label = 'loss',color = 'red')
plt.plot(r.history['val_accuracy'],label = 'validation_accuarcy',color = 'blue')
plt.legend()
plt.show()
```

```python
prediction = model.predict(X_test_mm)
```

```python
prediction = (prediction > 0.5)
```

```python
print(metrics.classification_report(y_test, prediction, target_names = ['Covid_Negative','Covid_Positive']))
```
